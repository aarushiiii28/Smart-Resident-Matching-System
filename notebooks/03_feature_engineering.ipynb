## This notebook is - Feature generator

# Load pairwise dataset

```python
import pandas as pd
import numpy as np

from pathlib import Path

DATA_PATH = Path("C:/Users/aarus_gzbjvu0/OneDrive/Desktop/hostel_pg/Roommate_compatibility/data/pairwise/pairwise_compatibility.csv")

df = pd.read_csv(DATA_PATH)
print(df.shape)
df.head()
```

# Difference Features
The model doesn't care much about:

Student A studies 7 hours
Student B studies 2 hours

It cares about:

Difference = 5

```python
difference_features = [

    "sleep_time",

    "wake_time",

    "study_hours",

    "cleanliness_score",

    "privacy_preference",

    "talkativeness",

    "friendship_expectation",

    "gaming_hours",

    "guest_frequency",

    "noise_sleep_tolerance"
]

for feature in difference_features:

    df[f"{feature}_diff"] = abs(df[f"{feature}_1"] - df[f"{feature}_2"])
```

# Similarity Features

```python
max_ranges = {

    "sleep_time": 12,

    "wake_time": 12,

    "study_hours": 6,

    "cleanliness_score": 4,

    "privacy_preference": 4,

    "talkativeness": 4,

    "friendship_expectation": 4,

    "gaming_hours": 8,

    "guest_frequency": 4,

    "noise_sleep_tolerance": 4
}

for feature in difference_features:

    max_diff = max_ranges[feature]

    df[f"{feature}_similarity"] = (

        1
        - (
            df[f"{feature}_diff"]
            / max_diff
        )
    )
```

# Aggregate Compatibility Features

```python
df["sleep_compatibility"] = ( df["sleep_time_similarity"] + df["wake_time_similarity"]) / 2
```

```python
df["study_compatibility"] = (

    df["study_hours_similarity"]

)
```

```python
df["cleanliness_compatibility"] = (

    df["cleanliness_score_similarity"]

)
```

```python
df["social_compatibility"] = (

    df["talkativeness_similarity"]

    + df["friendship_expectation_similarity"]

    + df["guest_frequency_similarity"]

) / 3
```

```python
df["privacy_compatibility"] = (

    df["privacy_preference_similarity"]

)
```

# Archetype Features

```python
df["same_lifestyle_type"] = (

    df["lifestyle_type_1"]

    ==

    df["lifestyle_type_2"]

).astype(int)
```

# Composite Compatibility Score

```python
df["behavioral_alignment_score"] = (

    df["sleep_compatibility"]

    + df["study_compatibility"]

    + df["cleanliness_compatibility"]

    + df["social_compatibility"]

    + df["privacy_compatibility"]

) / 5
```

# Save Engineered Dataset

```python
OUTPUT_PATH = Path(
    "../data/processed/pairwise_engineered.csv"
)

OUTPUT_PATH.parent.mkdir(
    parents=True,
    exist_ok=True
)

df.to_csv(
    OUTPUT_PATH,
    index=False
)

print(
    f"Saved to: {OUTPUT_PATH}"
)
```
