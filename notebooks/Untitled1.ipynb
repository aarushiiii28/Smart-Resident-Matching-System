### Model Explainability & Insights

* Feature importance analysis revealed that **behavioral alignment score** was the strongest predictor of roommate compatibility, indicating that overall lifestyle similarity plays a critical role in successful roommate matching.

* Students sharing the **same lifestyle archetype** (e.g., disciplined, social, night owl) showed significantly higher compatibility scores, highlighting the importance of behavioral patterns over demographic characteristics.

* **Social compatibility** emerged as one of the most influential factors, suggesting that similarities in social interaction preferences contribute substantially to harmonious co-living experiences.

* Differences in **sleep schedules** were among the strongest indicators of incompatibility, confirming that daily routine alignment is essential for reducing roommate conflicts.

* **Noise tolerance mismatch** was identified as a major source of incompatibility, demonstrating that environmental preferences strongly affect roommate satisfaction.

* Similarity in **cleanliness habits** significantly improved compatibility predictions, indicating that shared expectations regarding personal and shared spaces are important for successful roommate relationships.

* **Privacy preference alignment** contributed positively to compatibility, showing that students with similar expectations regarding personal space tend to coexist more effectively.

* Engineered behavioral features consistently outperformed raw demographic attributes such as age, course, and hostel block, validating the effectiveness of domain-driven feature engineering.

* Comparative model evaluation showed that **XGBoost outperformed Random Forest**, improving classification accuracy from **77.85% to 84.60%** and weighted F1-score from **0.776 to 0.847**, demonstrating the value of advanced ensemble learning techniques.

* The results suggest that roommate compatibility is primarily driven by **behavioral, lifestyle, and preference-based factors** rather than demographic similarities, providing actionable insights for intelligent roommate recommendation systems.
