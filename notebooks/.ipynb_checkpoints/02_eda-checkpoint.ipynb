# Correlation Analysis
We first analyze numerical relationships

```python
numerical_df = df.select_dtypes(
    include=[
        "int64",
        "float64"
    ]
)

numerical_df.head()
```

```python
### Correlation Matrix

correlation_matrix = numerical_df.corr()

correlation_matrix
```

```python
# Heatmap
plt.figure(figsize=(20,14))

sns.heatmap(
    correlation_matrix,
    cmap="coolwarm",
    center=0
)

plt.title(
    "Feature Correlation Heatmap"
)
plt.show()
```

# Strongest Correlations
Extracting most meaningful relationships.

```python
# Flatten Correlations
corr_pairs = correlation_matrix.unstack()
```

```python
# Remove Self-Correlation
corr_pairs = corr_pairs[
    corr_pairs != 1
]
```

```python
# Sort Correlations
sorted_corr = corr_pairs.sort_values(
    ascending=False
)

sorted_corr.head(20)
```

# Clustering
Clustering is an unsupervised machine learning technique that automatically organizes unlabeled data into natural groups based on their shared traits or patterns

### Feature selection
Feature selection is the process of identifying and choosing the most relevant variables (features) from a dataset to train a machine learning model. Its primary goal is to remove irrelevant, redundant, or noisy data, which helps improve model accuracy, reduce training time, and prevent overfitting.

```python
cluster_features = [
     "sleep_time",

    "wake_time",

    "study_hours",

    "cleanliness_score",

    "privacy_preference",

    "talkativeness",

    "gaming_hours",

    "gym_frequency"
]
```

### Scaling
Feature scaling is a crucial data preprocessing technique that transforms numerical features into a common scale or range. It prevents variables with larger magnitudes from disproportionately dominating the model, ensuring algorithms learn effectively, converge faster, and perform accurately.

```python
#StandardScaler is a tool in sklearn.preprocessing that standardizes dataset 
#features to a mean of 0 and a standard deviation of 1. 
#It shifts and scales each numeric column independently by computing the mathematical Z-score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_features = scaler.fit_transform(df[cluster_features])
```

### KMeans Clustering
https://www.youtube.com/watch?v=5FpsGnkbEpM

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters = 4, # tells the algorithm to divide the data into 4 groups (clusters)
    random_state = 42
)

clusters = kmeans.fit_predict(
    scaled_features
)
```

```python
# Add Cluster Labels
df["cluster"] = clusters

# Cluster Counts
df["cluster"].value_counts()
```

# PCA Visualization
https://www.youtube.com/watch?v=Dv-Kk7PDEas

```python
# PCA Reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_features = pca.fit_transform(
    scaled_features
)
```

```python
# Creating PCA df
pca_df = pd.DataFrame({
    "PCA1": pca_features[:, 0],

    "PCA2": pca_features[:, 1],

    "cluster": clusters
})
```

```python
plt.figure(figsize=(10,7))

sns.scatterplot(
    data = pca_df,
    x="PCA1",
    y="PCA2",
    hue="cluster",
    palette="Set2"
)
plt.title(
    "Student Behavioral Clusters"
)
plt.show()
```

# Silhouette Score
This measure: How well-separated clusters are.
#### Higher score: 
* better separation
* more meaninggul clustering
  
  https://www.youtube.com/watch?v=_jg1UFoef1c

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_avg = silhouette_score(
    scaled_features,
    clusters
)

print(
    "Silhouette Score: ",
    round(silhouette_avg, 3)
)
```

The silhouette score is low but it is still realistic/usable, because our dataset contains human behaviours and humans naturally overlap. Eg: A gamer can also be disciplined, an introvert can still be social sometimes etc. 

But still, we may remove noise. Example: bathing frequency, towel management, expense responsibility etc.

# Elbow Method

We determine best number of clusters instead of randomly choosing 4

```python
# Compute inertia
inertia_values = []

cluster_range = range(1,11)

for k in cluster_range:
    kmeans = KMeans(
        n_clusters = k,
        random_state=42
    )

    kmeans.fit(scaled_features)

    inertia_values.append(
        kmeans.inertia_
    )
```

```python
# Visualization
plt.figure(figsize=(8,5))

plt.plot(
    cluster_range,
    inertia_values,
    marker="o"
)

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.show()
```

The elbow method suggested that 4–5 clusters
provide a reasonable balance between compactness
and behavioral interpretability.

# DBSCAN
Unlike KMeans: 
* No fixed cluster count needed
* detects outliers
* density-based clustering

https://www.youtube.com/watch?v=RDZUdRSDOok
  

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(
    eps=1.6,
    min_samples=5
)

dbscan_clusters = dbscan.fit_predict(
    scaled_features
)
# eps - maximum distance for nearby points

#Imagine each student has a circle around them.
# eps = radius of that circle
# small eps → tiny circles → hard to connect points
# large eps → huge circles → almost everyone connects

# min_samples - minimum nearby points required to form a cluster

# So:
# if a point has enough nearby neighbors
# it becomes part of a dense cluster.
```

```python
df["dbscan_cluster"] = dbscan_clusters
```

```python
df["dbscan_cluster"].value_counts()
```

# t-SNE Visualization
More advanced than PCA

```python
from sklearn.manifold import TSNE
```

```python
# Reduce Dimensions

tsne = TSNE(
    n_components=2,
    random_state=42,
    perplexity=30
)

tsne_features = tsne.fit_transform(
    scaled_features
)
```

```python
tsne_df= pd.DataFrame({
    "TSNE1": tsne_features[:, 0],
    "TSNE2": tsne_features[:, 1],
    "cluster": clusters
})
```

```python
plt.figure(figsize=(10, 7))

sns.scatterplot(

    data=tsne_df,

    x="TSNE1",

    y="TSNE2",

    hue="cluster",

    palette="Set2"
)

plt.title(
    "t-SNE Behavioral Clusters"
)

plt.show()
```
